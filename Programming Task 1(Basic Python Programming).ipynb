```python
# Q2. Given an integer, n, perform the following conditional actions:

# If n is odd, print Weird
# If n is even and n in the inclusive range of  2 to 5 , print Not Weird
# If n is even and in the inclusive range of  6 to 20 , print Weird
# If n is even and greater than , print Not Weird

n = int(input())
if n %2 !=0:
    print('Weird')
else:
    if n >=2 and n <= 5:
        print('Not Weird')
    elif n>=6 and n <= 20:
        print('Weird')
    else:
        print('Not Weird')
```

```python
# Q6. Given a year, determine whether it is a leap year. If it is a leap year, return the Boolean True, otherwise return False.

    #Note that the code stub provided reads from STDIN and passes arguments to the is_leap function. It is only necessary to complete the is_leap function.
def is_leap(year):
    return (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0)

year = int(input())
print(is_leap(year))
```
