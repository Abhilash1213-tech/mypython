```python
# Q1. ven an array nums. We define a running sum of an array as runningSum[i] = sum(nums[0]â€¦nums[i]).Return the running sum of nums.
class Solution(object):
    def runningSum(self, nums):
        sum = 0
        k=[]
        a=len(nums)
        for i in range(a):
            sum += nums[i]
            k.append(sum)
        return k
a=Solution()
print(a.runningSum([1,2,3,4]))
```

```python
# Q2. Given the array nums consisting of 2n elements in the form [x1,x2,...,xn,y1,y2,...,yn]. Return the array in the form [x1,y1,x2,y2,...,xn,yn].

class Solution(object):
    def shuffle(self, nums, n):
        res = []
        for i in range(n):
            res.append(nums[i])
            res.append(nums[n + i])
        return res
a=Solution()
print(a.shuffle([2,5,1,3,4,7],3))
```

```python
# Q3. There are n kids with candies. You are given an integer array candies, where each candies[i] represents the number of candies the ith kid has, and an integer extraCandies, denoting the number of extra candies that you have.
# Return a boolean array result of length n, where result[i] is true if, after giving the ith kid all the extraCandies, they will have the greatest number of candies among all the kids, or false otherwise.


class Solution(object):
    def kidsWithCandies(self, candies, extraCandies):
        max_candies = max(candies)
        return [(c + extraCandies) >= max_candies for c in candies]

a=Solution()
print(a.kidsWithCandies([2,3,5,1,3],3))
```
