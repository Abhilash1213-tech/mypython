### Defanging an IP Address

```python
class Solution:
    def defangIPaddr(self, address: str):
        return address.replace('.','[.]')
s=Solution()
s.defangIPaddr("1.1.1.1")
```

### Find Numbers with Even Number of Digits

```python
class Solution:
    def findNumbers(self, nums: List[int]):
        count = 0
        for i in nums:
            i = str(i)
            if len(i)%2==0:
                count+=1
        return count                

s = Solution()
s.findNumbers([12,345,2,6,7896])
```

### Number of Good Pairs

```python
class Solution:
    def numIdenticalPairs(self, nums: List[int]):
        count = 0
        for i in range(len(nums)):
            for j in range(i+1,len(nums)):
                if nums[i]== nums[j]:
                    count += 1
        return count            

s = Solution()
s.numIdenticalPairs([1,2,3,1,1,3])
```

### How Many Numbers Are Smaller Than the Current Number

```python
from typing import List
class Solution:
    def smallerNumbersThanCurrent(self, nums: List[int]) -> List[int]:
        res = []
        for i in range(len(nums)):
            count = 0
            for j in range(len(nums)):
                if nums[j] < nums[i]:
                    count += 1
            res.append(count)
        return res

s = Solution()
s.smallerNumbersThanCurrent(nums=[8,1,2,2,3])
```

### Subtract the Product and Sum of Digits of an Integer

```python
class Solution:
    def subtractProductAndSum(self, n):
        sum = 0
        product = 1

        for i in range(len(str(n))):
            digit = n % 10
            sum = sum + digit
            product = product * digit
            n = n // 10   # â— remove last digit

        return product - sum


s = Solution()
print(s.subtractProductAndSum(234))

        
```

### XOR Operation in an Array

```python
class Solution:
    def xorOperation(self, n, start):
        result = 0

        for i in range(n):
            result ^= (start + 2 * i)

        return result
s = Solution()
s.xorOperation(5,0)
```
